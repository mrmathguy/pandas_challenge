```python
import pandas as pd


```

```python
student_data = pd.read_csv("students_complete.csv")

```

```python
school_data= pd.read_csv("schools_complete.csv")

```

```python
passing_grade=69
unique_schools = len(school_data["school_name"].unique())
total_students = len(student_data["Student ID"].unique())
total_budget = school_data["budget"].sum()
passing_math =len(student_data.loc[student_data["math_score"]>passing_grade,:])/len(student_data)
passing_reading =len(student_data.loc[student_data["reading_score"]>passing_grade,:])/len(student_data)
overall_passing = len(student_data.loc[(student_data["reading_score"]>passing_grade) & (student_data["math_score"]>passing_grade),:])/len(student_data)
avg_math = student_data["math_score"].mean()
avg_reading = student_data["reading_score"].mean()
```

```python
district_summary =pd.DataFrame({"Number of Schools":[unique_schools], 
                               "Number of Students":[total_students], 
                               "Total budget": [total_budget], 
                               "Average Math Score":[avg_math],
                                "Average Reading Score": [avg_reading],
                               "% of passing math": [(passing_math*100)], 
                               "% of passing reading":[(passing_reading*100)], 
                               "% of overall passing":[(overall_passing*100)]})
```

```python
district_summary
```

```python
total_data = pd.merge(student_data, school_data, on="school_name")
```

```python
total_data
```

```python
per_school_summary =total_data.groupby(["school_name","type"]).aggregate({"Student ID":"count", "budget": "mean", 
                                                                      "math_score":"mean", "reading_score":"mean"})
```

```python
per_school_summary
```

```python
passing_math =(total_data.loc[total_data["math_score"]>69,:])
```

```python
percent_passing_math = passing_math.groupby(["school_name","type"]).aggregate({"math_score":"count"})
```

```python
percent_passing_math["% passing math"] = 100*percent_passing_math["math_score"]/per_school_summary["Student ID"]
```

```python
percent_passing_math
```

```python
passing_reading = total_data.loc[total_data["reading_score"]>69,:]
```

```python
percent_passing_reading= passing_reading.groupby(["school_name", "type"]).aggregate({"reading_score":"count"})
```

```python
percent_passing_reading["% passing reading"] = 100*percent_passing_reading["reading_score"]/per_school_summary["Student ID"]
```

```python
percent_passing_reading.drop(columns= ["reading_score"])
```

```python
overall_passing = total_data.loc[(total_data["reading_score"]>69) & (total_data["math_score"]>69),:]
```

```python
percent_overall_passing = overall_passing.groupby(["school_name", "type"]).aggregate({"reading_score":"count"
                                                                                            })
```

```python
percent_overall_passing["% of overall passing"] = 100*percent_overall_passing["reading_score"]/per_school_summary["Student ID"]
```

```python
percent_overall_passing=percent_overall_passing.drop(columns =["reading_score"])
```

```python
percent_passing_math
```

```python
math_and_reading =pd.merge(percent_passing_math, percent_passing_reading, left_index=True, right_index=True)
```

```python
percent_overall_passing = pd.merge(math_and_reading, percent_overall_passing, left_index=True, right_index=True)
```

```python
percent_overall_passing=percent_overall_passing.drop(columns=["math_score", "reading_score"])
percent_overall_passing
```

```python
per_school_summary=per_school_summary.drop(columns=["math_score", "reading_score","Student ID"])
per_school_summary["Per Student Budget"]=per_school_summary["budget"]/per_school_summary["Total Students"]
per_school_summary
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_summary["Per Student Budget"], bins=spending_bins, labels=labels)
```

```python
per_school_summary
```

```python
spending_math_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"])["Average Math Scores"].mean()
spending_reading_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"])["Average Reading Scores"].mean()
spending_passing_math = per_school_summary.groupby(["Spending Ranges (Per Student)"])["% passing math"].mean()
spending_passing_reading = per_school_summary.groupby(["Spending Ranges (Per Student)"])["% passing reading"].mean()
overall_passing_spending = per_school_summary.groupby(["Spending Ranges (Per Student)"])["% of overall passing"].mean()
```

```python
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
per_school_summary["size"] = pd.cut(per_school_summary["Total Students"], bins=size_bins, labels= size_labels)
per_school_summary
```

```python
size_math_scores = per_school_summary.groupby(["size"])["Average Math Scores"].mean()
size_reading_scores = per_school_summary.groupby(["size"])["Average Reading Scores"].mean()
size_passing_math = per_school_summary.groupby(["size"])["% passing math"].mean()
size_passing_reading = per_school_summary.groupby(["size"])["% passing reading"].mean()
overall_passing_size = per_school_summary.groupby(["size"])["% of overall passing"].mean()
```

```python
size_summary=pd.DataFrame({"Mean Math Scores By Size":size_math_scores,
                          "Mean Reading Scores By Size": size_reading_scores,
                          "Passing Math By Size": size_passing_math,
                          "Passing Reading By Size": size_passing_reading,
                          "Overall Passing By Size":overall_passing_size })
```

```python
size_summary
```

```python
type_math_scores = per_school_summary.groupby(["type"])["Average Math Scores"].mean()
type_reading_scores = per_school_summary.groupby(["type"])["Average Reading Scores"].mean()
type_passing_math = per_school_summary.groupby(["type"])["% passing math"].mean()
type_passing_reading = per_school_summary.groupby(["type"])["% passing reading"].mean()
overall_passing_type = per_school_summary.groupby(["type"])["% of overall passing"].mean()
```

```python
type_summary =pd.DataFrame({"Mean Math Scores By type":type_math_scores,
                          "Mean Reading Scores By type": type_reading_scores,
                          "Passing Math By type": type_passing_math,
                          "Passing Reading By type": type_passing_reading,
                          "Overall Passing By type":overall_passing_type })
```
